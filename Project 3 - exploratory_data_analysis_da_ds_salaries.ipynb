```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('DataAnalystScientistFULL_cleaned.csv')

df.head(5)
```

```python
def simple_title(title):
    if 'manager' in title.lower():
        return 'manager'
    elif 'director' in title.lower():
        return 'director'
    elif 'data scientist' in title.lower():
        return 'data scientist'
    elif 'machine learning' in title.lower():
        return 'mle'
    elif 'data engineer' in title.lower():
        return 'data engineer'
    elif 'data consult' in title.lower():
        return 'data consultant'
    elif 'data architect' in title.lower():
        return 'data architect'
    elif 'analyst' in title.lower() and 'data scien' in title.lower():
        return 'data analyst/data scientist'    
    elif 'analyst' in title.lower():
        return 'data analyst'
    else:
        return 'na'
    
def job_level(title):
    if 'jr' in title.lower() or 'jr.' in title.lower() or 'junior' in title.lower():
        return 'junior'
    elif 'sr' in title.lower() or 'sr.' in title.lower() or 'senior' in title.lower():
        return 'senior'
    elif 'lead' in title.lower():
        return 'lead'
    elif 'principal' in title.lower():
        return 'principal'
    else:
        return 'na'
```

```python
df['simplified_title'] = df['Job Title'].apply(simple_title)
df['job_level'] = df['Job Title'].apply(job_level)
```

```python
df_categorized = df[['Location', 'Headquarters', 'Size', 'Type of ownership', 'Industry', 'Sector', 'Revenue', 'company_name', 'job_state', 'job_city', 'python', 'r studio', 'cloud', 'SQL', 'AWS', 'simplified_title',
       'job_level']]
df_cat_analysts = df_categorized[df_categorized.simplified_title == 'data analyst']
df_cat_scientists = df_categorized[df_categorized.simplified_title == 'data scientist']
```

```python
for i in df_categorized.columns:
    category_num = df_categorized[i].value_counts()
    print('Graph for %s: total = %d' % (i, len(category_num)))
    chart = sns.barplot(x=category_num.index, y=category_num)
    chart.set_xticklabels(chart.get_xticklabels(), rotation = 90)
    plt.show()
```
